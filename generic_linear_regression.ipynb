```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
data=datasets.load_boston()
input=data["data"]
output=data["target"]
input=pd.DataFrame(input)
output=pd.DataFrame(output)
#adding 1 to last column of input
number_of_features=len(input[0])
input["x (n+1)"]=1
#convert to numpy
input=np.array(input)
output=np.array(output)
print(input)
print(output)
```

```python
y=output.reshape(-1,1)
Y=np.asmatrix(y)
X=np.asmatrix(input)
X_transpose=np.transpose(X)
print("Y : ",Y.shape)
print("X : ",X.shape)
print("X_transpose : ",X_transpose.shape)
```

```python
#so ab tak sab sahi
X_into_X_transpose_ka_inverse=np.linalg.inv(X*X_transpose)
coefficient_array=X_transpose*X_into_X_transpose_ka_inverse*Y
print(coefficient_array)
```

```python
code_se_coef_=[]
for i in range (len(coefficient_array)-1):
    code_se_coef_.append(coefficient_array[i][0])
print(code_se_coef_)
```

```python
#so yeh matrix aayega
#change to list
coef_=coefficient_array.tolist()
print(coef_)
```

```python
code_se_coef_=[]
for i in range (len(coef_)-1):
    code_se_coef_.append(coef_[i][0])
print(code_se_coef_)
```

```python
code_se_intercept_=coef_[len(coef_)-1][0]
print(code_se_intercept_)
```
